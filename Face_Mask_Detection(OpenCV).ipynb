```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from keras.preprocessing.image import img_to_array,load_img
```

```python
model=load_model('facemask_model.h5')
```

```python
img_width,img_height=150,150
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap=cv2.VideoCapture(0)
img_count_full=0
font=cv2.FONT_HERSHEY_SIMPLEX
org=(1,1)
class_label=''
fontScale=1
color=(255,0,0)
thickness=2

while True:
    src,img=cap.read()
    if src==False:
        break
    scale=50
    width=int(img.shape[1])
    height=int(img.shape[0])
    dim=(width,height)
    
    img=cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray_img,1.1,6)
    
    img_count=0
    
    for (x,y,w,h) in faces:
        org=(x-10,y-10)
        img_count+=1
        color_face=img[y:y+h,x:x+w]
        cv2.imwrite('input.jpg',color_face)
        l_image=load_img('input.jpg',target_size=(img_width,img_height))
        l_image=img_to_array(l_image)
        l_image=np.expand_dims(l_image,axis=0)
        print(l_image.shape)
        prediction=model.predict(l_image)
        
        if prediction==0:
            class_label="MASK"
            color=(0,255,0)
        else:
            class_label="NO MASK"
            color=(0,0,255)
        
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        cv2.putText(img,class_label,org,font,fontScale,color,thickness,cv2.LINE_AA)
        
    cv2.imshow('Face Mask Detection',img)
    if cv2.waitKey(1)& 0xFF==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```
